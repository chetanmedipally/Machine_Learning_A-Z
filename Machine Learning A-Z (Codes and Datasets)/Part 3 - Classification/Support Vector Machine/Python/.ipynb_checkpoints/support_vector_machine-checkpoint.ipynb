# Support Vector Machine (SVM)

## Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
print(X)
```

```python
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scalar_X = StandardScaler()
X_train = scalar_X.fit_transform(X_train)
X_test = scalar_X.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the SVM model on the Training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear',random_state=0)
classifier.fit(X=X_train,y=y_train)
```

## Predicting a new result

```python
print(classifier.predict(scalar_X.transform([[30,87000]])))
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
#print(y_pred.reshape(len(y_pred),1))
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),axis=1))
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_true=y_test,y_pred=y_pred)
print(cm)
accuracy_score(y_true=y_test,y_pred=y_pred)
```

## Visualising the Training set results

```python
from matplotlib.colors import ListedColormap
X_set,y_set = scalar_X.inverse_transform(X_train),y_train
X1,X2 = np.meshgrid(np.arange(start = X_set[:,0].min() - 10,stop = X_set[:,0].max() + 10,step = 0.25),
                    np.arange(start = X_set[:,1].min() - 1000,stop = X_set[:,1].max() + 1000,step = 0.25))
plt.contourf(X1,X2,classifier.predict(scalar_X.transform(np.array([X1.ravel(),X2.ravel()]).T)).reshape(X1.shape),
            alpha = 0.75,cmap = ListedColormap(('red','green')))
plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set==j,0],X_set[y_set==j,1],c=ListedColormap(('red','green'))(i),Label = j)
plt.title('Support Vector Machines (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```python
from matplotlib.colors import ListedColormap
X_set,y_set = scalar_X.inverse_transform(X_test),y_test
X1,X2 = np.meshgrid(np.arange(start = X_set[:,0].min() - 10,stop = X_set[:,0].max() + 10,step = 0.25),
                    np.arange(start = X_set[:,1].min() - 1000,stop = X_set[:,1].max() + 1000,step = 0.25))
plt.contourf(X1,X2,classifier.predict(scalar_X.transform(np.array([X1.ravel(),X2.ravel()]).T)).reshape(X1.shape),
            alpha = 0.75,cmap = ListedColormap(('red','green')))
plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set==j,0],X_set[y_set==j,1],c=ListedColormap(('red','green'))(i),Label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
